```python
# this is a wakthrough of the code in https://gym.openai.com/docs/
import gym
```

```python
env = gym.make('CartPole-v0')
```

```python
print(env.action_space) # action_space and observation_space are of type Space
```

Discrete space allows a fixed range of non-negative integers - Here, it can be 0 or 1

```python
print(env.observation_space)
```

Box allows for n-dimensional possible observation spaces. Here, it will be an array of 4-dimensions

```python
env.observation_space.high # the ends of possible values that can be taken by 4-dimensions of Box
```

```python
env.observation_space.low
```

```python
env.action_space.high # no  such attributes for action_space, because it is discrete in nature
```

```python
for _ in range(20): # 20 lives/episodes
    observation = env.reset()
    for i in range(1000):
        env.render()
        print("Step {}: Observation: {}".format(i+1, observation))
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("The episode has terminated after {} steps".format(i + 1))
            break
        
```

```python
# experimenting with spaces module
from gym import spaces
some_env = spaces.Discrete(8)  # contains set [0, 1, 2, 3, 4 ...6, 7]
x = some_env.sample()
assert some_env.contains(x)
assert some_env.n == 8   # Size of a discrete set is referred to by 'n', NOT by 'size'
```
